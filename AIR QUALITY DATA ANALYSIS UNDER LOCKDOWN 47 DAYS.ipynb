```python
import pandas as pd
data = pd.read_csv('iti-shahdra, jhilmil industrial area, delhi-air-quality.csv')
data.head(20)
```

```python
data.columns
```

```python
# remove space between names
data = data.rename(columns={
    " pm25":"pm25",
    " pm10":"pm10",
    " o3":"o3",
    " no2":"no2",
    " so2":"so2",
    " co":"co"
})
```

```python
data.columns
```

```python
# extract date of lockdown

data['date'] = pd.to_datetime(data.date)
data['date']
```

```python
day47 = data.loc[data['date']>'2020-03-24']
day47= day47.sort_values(by='date')
```

```python
day47.replace(' ',0,inplace=True)
day47
```

```python
import matplotlib.pyplot as plt
```

```python
#convert data frame into list
```

```python
dates = day47['date']
pm25 = day47['pm25']
pm25 =[ int(i) for i in pm25]
plt.figure(figsize=(10,8))
plt.plot(dates,pm25)
plt.title("pm 2.5 values under lockdown")
plt.xlabel(" lockdown dates")
plt.ylabel("pm 2.5 values")
plt.show()
```

```python
# between 8 feb and 10 may
mask =   (data['date']>='2020-02-07') & (data['date']< '2020-03-25')
beforelockdown = data.loc[mask]
beforelockdown.replace(' ',0,inplace=True)
beforelockdown.info()
```

```python
bl_pm25= beforelockdown['pm25']
bl_pm25 = [int(i) for i in bl_pm25]
plt.figure(figsize=(10,8))
length = [int(i) for i in range(1,len(dates)+1)]
plt.title("Before lockdown vs under lockdown pm2.5 values")

plt.plot(length,pm25,color='blue',label="under lockdown")
plt.plot(length,bl_pm25,color='red',label ="before lockdown")
plt.legend()
plt.show()
```
